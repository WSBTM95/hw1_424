```python
df_c = df_raw.copy()
df_c['Date'] = df_raw["Date"].map(lambda x: datetime.strftime(x, '%b'))
#df_cleaning['Date'] =  df_raw["Date"].map(lambda x: datetime.strftime(x, '%b'))
#df_cleaning
#df_cd = df_c.rename(columns={"Date":"Month"})#[(((df_c['Year']==2019) or (df_c["Year"]==2020)) and (df_c["Month"]=="Apr"))]
```

```python
df_cd = df_c.rename(columns={"Date":"Month"})
y19 = (df_cd["Year"] == 2019)
y20 = (df_cd['Year']==2020)
apr = (df_cd["Month"]=="Apr")
dff = df_cd[["Year", "Month", "Primary Type", "ID"]]
dff
```

```python
dff = dff[(y19|y20)&apr].groupby(["Year", "Month", 'Primary Type']).count().reset_index()
dff = pd.pivot_table(dff, values='ID', index=['Primary Type'], columns=['Year'])
```

```python
dff = dff.fillna(0)
dy19 = dff[2019].to_numpy()
dy20 = dff[2020].to_numpy()
dyc = []
for x in range(0, len(dy19)):
    dyc.append(round((dy20[x]/dy19[x] - 1)*100, 1))
dff["Change"] = dyc
dff.sort_values(by=["Change"])
```

```python
dfm = df_raw.copy()
dfm = dfm[["Date", "ID"]]
dfm["Date"] = dfm["Date"].map(lambda x: datetime.strftime(x, '%b'))
dfm = dfm.rename(columns={"Date":"month"}).groupby(["month"]).count()
dfm
```

```python
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", 
          "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
dm = dfm.copy()
dm = dm.reset_index()
dm['month'] = pd.Categorical(dm['month'], categories=months, ordered=True)
dm = dm.sort_values(by=["month"])
dm['ID'] = dm["ID"].map(lambda x: round(x/100000, 2))
dm
```

```python
dft = df_raw.copy()[["Date", "ID", "Year"]]
dft['Date'] = dft["Date"].map(lambda x: datetime.strftime(x, '%H'))
dft = dft.groupby(["Year", "Date"]).count().rename(columns={"Date":"Hr", "ID":"count"})
dft
```
